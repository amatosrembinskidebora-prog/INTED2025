4 - Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual faixa etária está à idade:

até 12 - Você é uma criança!

de 13 à 18 - Você é um Adolescente!

de 19 à 60 - Você é um Adulto!

acima de 60 - Você é idoso!

```python
idade = int(input("Digite sua idade: "))

if (idade <= 12):
  print("Você é uma criança!")
elif (idade <= 18):
  print(" Você é um Adolescente!")
elif (idade <= 60):
  print (" Você é um Adulto!")
else:
  print("Você é idoso!")
```

```python
opcao = input("Digite 'S' para continuar ou 'N' para sair: ")

if opcao.upper() == "S":
  print("Você escolheu continuar.")
elif opcao.upper() == "N":
  print("Você escolheu sair.")
else:
  print("Opção inválida.")
```

Peça um número de 1 a 7 e informe o dia da semana correspondente. Crie um algoritmo em Python que associe o número ao dia correto.

```python
dia_numero = int(input("Digite um número de 1 a 7: "))

if dia_numero == 1:
  print("Domingo")
elif dia_numero == 2:
  print("Segunda-feira")
elif dia_numero == 3:
  print("Terça-feira")
elif dia_numero == 4:
  print("Quarta-feira")
elif dia_numero == 5:
  print("Quinta-feira")
elif dia_numero == 6:
  print("Sexta-feira")
elif dia_numero == 7:
  print("Sábado")
else:
  print("Número inválido. Digite um número de 1 a 7.")
```
